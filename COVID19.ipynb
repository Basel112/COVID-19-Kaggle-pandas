# DATA VISUALIZATION

```python
group_by_Date= dataset.groupby("Date").agg({"Confirmed":"sum","Recovered":"sum","Deaths":"sum"}).reset_index()
```

```python
group_by_Date
```

```python
def graph(X):
    
    x=list(X.index)
    plt.plot(x,X["Confirmed"])
    plt.plot(x,X["Recovered"])
    plt.plot(x,X["Deaths"])
    plt.legend(['Confirmed','Recovered','Deaths'],loc= 'upper left')
```

```python
x=list(group_by_Date.index)
graph(group_by_Date)
```

```python
group_by_Date.index=pd.to_datetime(group_by_Date["Date"]) ## index by date 

group_by_Month=group_by_Date.groupby(pd.Grouper(freq="M")).agg({"Confirmed":"max","Recovered":"max","Deaths":"max"}).reset_index()
```

```python
group_by_Month
```

```python
x=list(group_by_Month.index)
plt.xticks(x,group_by_Month["Date"])
graph(group_by_Month)
```
